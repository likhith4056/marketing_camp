```python
df=df.dropna()
print('after dropping missing values',df.shape)
```

```python
df=df.drop_duplicates()
print("after dropping duplicates",df.shape)
```

```python
df['Marital_Status']=df['Marital_Status'].str.lower().str.strip()
```

```python
print(df.columns.tolist())
```

```python
df.head()
```

```python
df['Education']=df['Education'].str.lower().str.strip()
```

```python
df.head()
```

```python
df.columns = [col.lower().replace(" ", "_") for col in df.columns]
```

```python
# Convert 'income' to float (if not already)
df['income'] = df['income'].astype(float)

# Create 'age' column from 'year_birth'
df['age'] = pd.Timestamp.now().year - df['year_birth']
```
